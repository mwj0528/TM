```python
# 1) 원하는 열 하나만 뽑아내기 (명칭(원문))
name=data.loc[:,['명칭(원문)']]
name.head()
```

```python
# 2) 조건에 맞는 열 여러개 뽑아내기 (출원이라는 단어가 포함된 열만)
apply = data.columns.str.contains("출원")
apply_data=data.loc[:,apply]
apply_data.head()
```

```python
# 3) 원하는 행 하나만 뽑아내기 (KR20200084076A)
num_28=data.loc[data['번호']=='KR20200084076A',:]
num_28
```

```python
# 4-1*) 조건에 맞는 행 여러개 뽑아내기 (ex, 출원인 국가가 KR, 출원인대표명이 한국기술교육대학교산학협력단, 출원일이 2021년도)
condition_1=data.loc[(data['출원인국가']=='KR') & (data['출원인대표명']=='한국기술교육대학교산학협력단') & (data['출원일'].str.contains('2021')),:]
condition_1
```

```python
# 4-2*) 출원일이 2020년도 이후이면서도, CPC(공통특허분류) 코드 중 G05B을 포함하는 특허만 뽑아내시오
condition_2=data.loc[(data['공통특허분류'].str.contains('G05B')) & (data['출원일']>='2020.01.01'),:]
condition_2.head()
```

```python
# 5) 조건에 맞는 행과 열 복합 조건으로 뽑아내기
df=data.loc[(data['공통특허분류'].str.contains('G05B')) & (data['출원일']>='2020.01.01') , ['번호','명칭(원문)']]
df.head()
```

```python
# 6) 두개의 열 합쳐서 새로운 열 만들기 (출원인 김선혜 + from + 출원인 국가를 합치기 > '김선혜 from US')
data['출원인+출원인 국가']=data['출원인대표명']+" from "+data['출원인국가']
data['출원인+출원인 국가'].head()
```

```python
# 7) 하나의 열에서 원하는 부분 뽑아내서 새로운 열 만들기 (등록일에서 등록 년도만 뽑아내기)
data['year']=data['출원일'].str[0:4]
data['year']=data['year'].astype(int)
data['year'].head()
```

```python
# 8) 특허 국제 특허 분류 처리하기 (리스트)
list=[]
for i in data['국제특허분류']:
    list.append(i.split(", "))
data['국제특허분류(리스트)']=list
data.head()
```

```python
# 9-1) 인용 정보 네트워크로 만들기 (번호 from node - 자국피인용특허to node) - 자국 피인용 있는 애들로
from_list=[]
to_list=[]
for i, key in enumerate(data['번호']):
    if pd.notna(data.loc[i, '자국피인용특허']):
        a=data.loc[i, "자국피인용특허"].split(",")
        
        for j in a:
            from_list.append(key)
            to_list.append(j)
df=pd.DataFrame(zip(from_list,to_list),columns=['번호','자국피인용특허'])
df
```

```python
# 9-2) 9-1에서 뽑아낸 특허들의 제목만 뽑아오기 (from 특허번호, from 특허 제목, to 특허번호, to 특허 제목) 네개의 열을 가진 dataframe 만들기
from_title=[]
to_title=[]
for i in from_list:
    for idx, j in enumerate(data['번호']):
        if i == j:
            from_title.append(data.loc[idx, "명칭(원문)"])
for l in to_list:
    c = 0    
    for index, k in enumerate(data['번호']):
        if l == k:
            c = 1
            to_title.append(data.loc[index, "명칭(원문)"])
        if c == 0:
            to_title.append(pd.NA)
            
df=pd.DataFrame(zip(from_list,from_title,to_list,to_title),columns=['특허 번호', '특허 제목','피인용특허 번호','피인용특허 제목'])
    
df
```
