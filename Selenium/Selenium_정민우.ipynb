```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By

url = "https://sports.news.naver.com/index"
driver = webdriver.Chrome()
driver.set_window_size(1920,1080)
driver.get(url)

time.sleep(1)

driver.find_element(By.XPATH,'//*[@id="lnb_list"]/li[5]/a').click()
time.sleep(1)
driver.find_element(By.XPATH,'//*[@id="content"]/div/div[1]/div[2]/div[1]/a').click()
time.sleep(1)
title = driver.find_elements(By.CLASS_NAME,'title')

for i in title:
    print(i.text)

driver.close()
```
