```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
```

```python
#1번 데이터 이용
data=pd.read_csv("C:/Users/Administrator/Desktop/2_network_analysis&prediction/1.csv")
```

```python
data.head()
```

```python
# 네트워크 그래프 생성
g = nx.Graph()
el = data.values
g.add_edges_from(el)
nx.draw(g,with_labels=True)
```

```python
print(type(g))
print(g.edges())
```

```python
#Degree에 따라 노드 크기를 표현한 그래프
degree=g.degree()
node_size=[100 + d[1]*100 for d in degree]
nx.draw(g,node_size=node_size, with_labels=True)
```

```python
#Closeness 계산
closeness = nx.closeness_centrality(g)
closeness
```

```python
#closeness가 클수록 진한 색
node_colors = list(closeness.values())

# 그래프 그리기
nx.draw(g, with_labels=True, node_size=node_size, node_color=node_colors, cmap=plt.cm.Spectral_r)
```

```python
#betweeness 계산
betweeness=nx.betweenness_centrality(g)
betweeness
```

```python
#betweeness가 클수록 진한 색
node_colors = list(betweeness.values())

# 그래프 그리기
nx.draw(g, with_labels=True, node_size=node_size, node_color=node_colors, cmap=plt.cm.Reds)
```

```python
#density
density=nx.density(g)
density
```

```python
#Cluster 
clusters = nx.connected_components(g)
print("Clusters:")
for c in clusters:
    print(len(c),end=" ")
```

```python
# 모든 clique 탐색
cliques = list(nx.find_cliques(g))

# 각 clique의 크기
clique_sizes = [len(clique) for clique in cliques]

# 가장 큰 clique 도출
largest_clique_index = clique_sizes.index(max(clique_sizes))
largest_clique = cliques[largest_clique_index]
largest_clique
```

```python
#largest_clique: 황금색, 나머지: 회색
colors = ["gold" if node in largest_clique else "grey" for node in g.nodes]

nx.draw(g, with_labels=True, node_color=colors, cmap=plt.cm.Blues, font_color="black")
```

```python
# 우선연결성 계산(각 노드의 연결선 수끼리의 곱 -> 더 많은 연결을 가진 노드일수록 미래에 더 연결될 확률이 높다)
predicted_edges = nx.preferential_attachment(g)

result_df = pd.DataFrame(predicted_edges, columns=['from', 'to', 'preferential_attachment'])

result_df['prob'] = result_df['preferential_attachment'] / result_df['preferential_attachment'].sum()

```

```python
# 정렬
result_df.sort_values(by='prob', ascending=False, inplace=True)
result_df.reset_index(drop=True, inplace=True)
```
